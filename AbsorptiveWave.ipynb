# Wave with Absorption
   
Plot the electric field of a wave in an absorptive dielectric.
The dielectric material is in the region $x>0$ and the region of negative $x$ is vacuum.
$$\frac{E}{E_0} = e^{-\kappa\omega x/c} \cos\left( \omega\left( \frac{nx}{c}-t \right) + \phi \right)$$
Take the constants from lecture: $\phi=0$, $n=1.5$, $\kappa=0.07$, and $f=5.0\times10^{14}\,\text{Hz}$.
Take $t=0$ as a snapshot of the wave.

In the calculation we use $dx=\dfrac{c}{\kappa\omega}$ as the distance over which the wave is attenuated by $e^{-1}$ called the skin depth, and $k=\dfrac{n\omega}{c}$ as the constant multiplying $x$ in the cosine function.  We calculate the envelope function $\text{env}=e^{-x/dx}$ that gives the time independent boundary of the time dependent part given by
$$\dfrac{E}{E_0} = \text{env}\cos(kx)$$
for $t=0$ and $\delta=0$.
The function is plotted from the interface ($x=0$) to $7\,\mu\text{m}$ of depth into the material.

```python
import numpy as np
import matplotlib.pyplot as plt

n=1.5
kappa=0.07
omega = 2.0*np.pi*5.0e14
dx = (3.0e8)/(kappa*omega)
k = n*omega/(3.0e8)
x = np.linspace(0,5.*dx,500)
env = np.exp(-x/dx)
E = env*np.cos(k*x)
plt.plot(x,env,linestyle="dashed",color="green",label='envelope')
plt.plot(x,-env,linestyle="dashed",color="green")
plt.plot(x,E,color="blue",label='wave')
plt.xlabel("$x$")
plt.ylabel("$E/E_0$")
plt.legend()

plt.show()
```
